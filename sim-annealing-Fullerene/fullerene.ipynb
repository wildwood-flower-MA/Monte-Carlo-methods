```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix
```

```python
def plot_molecule(
                filename_data: str,
                filename_savefig: str,
                elev: float = 0, azim: float = 0,
                
                r_max: float = 3,
                closest_neighbors: int = 3
                ) -> None:

    data = np.loadtxt(filename_data, usecols=(0,1,2), skiprows=0)
    data = data[-60:]
    #data = data[:60]
    #neighbors = np.argsort(distance_matrix(data, data), axis=1)[:, 1:closest_neighbors+1]

    dist_matrix = distance_matrix(data, data)
    neighbors = [np.where(dist_matrix[i] < r_max)[0] for i in range(len(data))]
    neighbors = [nbrs[nbrs != i] for i, nbrs in enumerate(neighbors)]

    edges = set()
    for i, nbrs in enumerate(neighbors):
        for j in nbrs:
                a, b = sorted((i, j))
                edges.add((a, b))

    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_box_aspect([1,1,1])

    ax.elev = elev
    ax.azim = azim

    viewer_direction = np.array([
        np.cos(np.radians(elev))*np.sin(np.radians(azim)),
        np.cos(np.radians(elev))*np.cos(np.radians(azim)),
        np.sin(np.radians(elev))
    ])

    for data_n in data:
        distance_to_viewer = np.dot(data_n, viewer_direction)
        alpha = max(0.1, min(1, 1 - distance_to_viewer/np.max(np.linalg.norm(data_n))))
        ax.scatter(data_n[0], data_n[1], data_n[2], color='k', alpha = alpha)

    for i, j in edges:

        x0, y0, z0 = data[i]
        x1, y1, z1 = data[j]

        distance_to_viewer = np.dot(data[i], viewer_direction)
        alpha = max(0.1, min(1, 1 - distance_to_viewer/np.max(np.linalg.norm(data[i]))))
        #ax.scatter(x0, y0, z0, color='k', alpha = alpha)

        midpoint = np.array([(x0 + x1)/2, (y0 + y1)/2, (z0 + z1)/2])
        distance_to_viewer = np.dot(midpoint, viewer_direction)
        alpha = max(0.1, min(1, 1 - distance_to_viewer/np.max(np.linalg.norm(data, axis=1))))
        ax.plot([x0, x1], [y0, y1], [z0, z1],
                color='black', linewidth=1, alpha=alpha)
        
    plt.tight_layout()
    plt.savefig(filename_savefig, dpi=300, bbox_inches='tight')
    plt.show()
```

```python
plot_molecule("atoms_positions_c60.dat", "c60_fullerene.pdf",
            r_max = 2.0, elev = 45, azim = 45)
```

```python
plot_molecule("zad2_vectors.dat", "zad2.pdf",
            r_max = 2.1, elev = 45, azim = 45)

molecule = np.loadtxt("zad2_vectors.dat", usecols=(0,1,2), skiprows=0)

fig, ax = plt.subplots(figsize = (4,4))
rs = np.linalg.norm(molecule, axis=1)
r_avg = np.mean(rs.reshape(-1, 60), axis=1)
ax.plot(r_avg, color = 'k')
ax.set_xlabel('iteracja/n')
ax.set_ylabel(r'$\overline{r}$')
plt.savefig("zad2fulleren.pdf")
plt.show()


data = np.loadtxt("zad2_pcf.dat")[-1]
hist, bin_edges = np.histogram(data, bins = 100)

fig, ax = plt.subplots(figsize = (4,4))
#ax.hist(data, bins=bin_edges, alpha=1, color = 'r', edgecolor='black')
ax.plot(np.linspace(0, 2.5*np.average(data), len(data)), data, color = 'k')
ax.set_ylabel('zliczenia')
ax.set_xlabel('wartość PCF')
fig.tight_layout()
plt.savefig("zad2PCF.pdf")
plt.show()

data = np.loadtxt("zad2_energy.dat")[:]
fig, ax = plt.subplots(figsize = (4,4))
linsp = np.linspace(0, len(data)*100, len(data))
ax.plot(linsp, data, color = 'k', label = f'E')
ax.plot(linsp, np.ones_like(linsp)*data[-1], '--',
        alpha = 0.8, color = 'r', label = f'E_min = {data[-1]}')
ax.set_ylabel('energia')
ax.set_xlabel('iteracja')
ax.grid()
ax.legend()
ax.set_ylim([-360, -300])
fig.tight_layout()
plt.savefig("zad2energia.pdf")
plt.show()
```
