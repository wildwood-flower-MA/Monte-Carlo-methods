```python
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as mgimg
import matplotlib.animation as animation
plt.rc('xtick', labelsize=10) 
plt.rc('ytick', labelsize=10)
```

```python
def f1(x: float):
    return 1 + np.tanh(x)

def f2(x: float):
    return 1.0/(1.0 + x**2)

def f3(x: float):
    return np.cos(np.pi*x)**10

lim_1 = (-3,3)
lim_2 = (0,10)
lim_3 = (0,1)
```

```python
data = [590, 1861, 5839, 14668, 26388, 26692, 15392, 5933, 2024, 607]
lim = lim_1
integral = 6.00156
variance = 0.00018
f = lambda x: f1(x)
```

```python
data = [50652, 30176, 10011, 4214, 2113, 1138, 688, 460, 317, 226]
lim = lim_2
integral = 1.47147
variance = 0.00065
f = lambda x: f2(x)
```

```python
data = [20512, 21347, 6081, 416, 1, 1, 387, 6372, 24740, 20140]
lim = lim_3
integral = 0.24588 
variance = 0.00077 
f = lambda x: f3(x)
```

```python
N = len(data)

fig, ax = plt.subplots(figsize = (5,5))
bin_edges = np.linspace(lim[0], lim[1], N + 1)
ax.bar(bin_edges[:-1], data,
        width=np.diff(bin_edges), align='edge', edgecolor='black')
linsp = np.linspace(lim[0], lim[1], 500)

ax.set_xlim(lim)
ax.set_xticks(bin_edges)
ax.set_xticklabels([f'{bin_edge:.3}' for bin_edge in bin_edges], rotation=45)
ax.set_xlabel('x', fontsize=10)
ax.set_ylabel('n', fontsize=10)

fprim = lambda x: abs(f(x+0.0001) - f(x))/0.0001
axt = ax.twinx()
axt.plot(linsp, [f(x) for x in linsp], color = 'k', label = r'$f$')
axt.plot(linsp, [fprim(x) for x in linsp], '--', color='k', label=r'$df/dx$')
axt.set_ylabel('f(x)', fontsize=10)
plt.title(f'wartość całki = {integral}, ' + r'błąd względny ' + f'= {variance}', fontsize=10)
plt.legend(fontsize=12, frameon=False)
plt.savefig('f.pdf', format = 'pdf')
plt.show()
```
